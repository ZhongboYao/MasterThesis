```python
def parse_questions(question_list):
    questions_dict = {}
    for question in question_list:
        # Extract the question text (before '\n')
        question_text = question.split('\n')[0].strip()
        # Remove numbering if present (e.g., "1. ")
        if question_text[1] == '.':
            question_text = question_text[3:].strip()
        
        # Extract the choices
        choices = []
        if 'Choose one answer from' in question:
            # Get the part after 'Choose one answer from'
            choices_str = question.split('Choose one answer from')[1].strip()
            # Remove brackets and split by commas
            choices = [choice.strip() for choice in choices_str.strip('[]').split(',')]
        elif 'Answer with' in question:
            # For open-ended questions, use the provided instruction
            choices = [question.split('Answer with')[1].strip()]
        
        questions_dict[question_text] = choices
    
    return questions_dict

# Input question list
question_list = [
    "1. Which jaw is affected by the lesion?\n Choose one answer from [Mandible, Maxilla, Mandible and maxilla]\n",
    "2. Where is the lesion centered anatomically?\n Choose one answer from: [Molar region, Ramus region, Incisor region, Sinus region, TMJ region, Canine/Premolar region]\n",
    "3. How does the lesion relate to the surrounding teeth?\n Choose one answer from: [Apex associated Vital tooth, Apex associated Non-vital tooth, Apex associated tooth with unknown vitality unknown, Root associated, Crown associated, Missing tooth associated, Not tooth associated]\n",
    "4. How many lesions are present?\n Choose one answer from: [1, 2, >=3, Generalised lesion]\n",
    "5. What is the maximum size of the lesion?\n Choose one answer from: [<2 cm, 2-3 cm, >3 cm]\n",
    "6. What is the anatomical origin of the lesion?\n Choose one answer from: [Central, Peripheral]\n",
    "7. Specify the characteristics of the borders of the lesion?\n Choose one answer from: [Corticated, Defined but not corticated, Diffuse]\n",
    "8. The loculation of the lesion is?\n Choose one answer from: [Unilocular, Multilocular, Not loculated]\n",
    "9. What is the radiographic appearance of the contents of the lesions?\n Choose one answer from: [Radiolucent, Radio-opaque, Mixed, Radiolucent with flecks, Opaque]\n",
    "10. Does the lesion include one or more teeth?\n Choose one answer from: [Yes, No]\n",
    "11. Is there expansion of the bony cortex?\n Choose one answer from: [Yes, No]\n",
    "12. Does the lesion result in root resorption?\n Choose one answer from: [Yes, No]\n",
    "13. Are there any signs of tooth displacement or impaction?\n Choose one answer from: [Yes, No]\n",
    "14. What is the most possible lesion given the above information?\n Answer with the best possible lesion name.\n"
]

# Get the dictionary
questions = parse_questions(question_list)

# Print the result
for question, answers in questions.items():
    print(f"{question}: {answers}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1) Load
df = pd.read_csv('evaluation_dataset/validation.csv')

# 2) Make the first column (questions) the index
q_col = df.columns[0]
df = df.set_index(q_col)

# 3) Select the last 14 questions
selected = [7, 8, 9, 10, 11, 12, 14, 15, 17, 18, 20, 21, 22, 24]
q_df = df.iloc[selected, :]

# 4) Plot
fig, axes = plt.subplots(5, 3, figsize=(15, 20))
axes = axes.flatten()

for idx, (question_text, row) in enumerate(q_df.iterrows()):
    counts = row.value_counts()
    ax = axes[idx]
    ax.bar(counts.index.astype(str), counts.values)
    ax.set_title(question_text, fontsize=12)

    # Rotate and right-align labels, reduce font size
    ax.set_xticklabels(counts.index.astype(str), rotation=45, ha='right', fontsize=8)
    ax.tick_params(axis='y', labelsize=8)

# Turn off any unused subplots
for ax in axes[len(q_df):]:
    ax.axis('off')

# Tighten layout and add more bottom padding
plt.tight_layout()
plt.subplots_adjust(bottom=0.2)  # give extra space for rotated labels
plt.show()
```

```python
import json
from collections import Counter, defaultdict

# 1. Load your JSON data (replace the path or variable as needed)
with open("evaluation_dataset/doc_anns.json", "r") as f:
    records = json.load(f)

# 2. Define the 14 question keys exactly as they appear in the JSON
questions = [
    "Which jaw is affected by the lesion?\n",
    "Where is the lesion centered anatomically?\n",
    "How does the lesion relate to the surrounding teeth?\n",
    "How many lesions are present?\n",
    "What is the maximum size of the lesion?\n",
    "What is the anatomical origin of the lesion?\n",
    "Specify the characteristics of the borders of the lesion?\n",
    "The loculation of the lesion is?\n",
    "What is the radiographic appearance of the contents of the lesions?\n",
    "Does the lesion include one or more teeth?\n",
    "Is there expansion of the bony cortex?\n",
    "Does the lesion result in root resorption?\n",
    "Are there any signs of tooth displacement or impaction?\n",
    "What is the most possible lesion given the above information?\n"
]

# 3. Tally counts
counts = defaultdict(Counter)
for case in records:
    for q in questions:
        answer = case.get(q, "").strip()
        if answer:
            counts[q][answer] += 1

# 4. Build the final dict (strip trailing newlines from questions)
choices_per_feature = {q.strip(): dict(counts[q]) for q in questions}

# 5. Write out as JSON
with open("choices_per_feature.json", "w") as out:
    json.dump(choices_per_feature, out, indent=2)

print("Generated choices_per_feature.json")
```
